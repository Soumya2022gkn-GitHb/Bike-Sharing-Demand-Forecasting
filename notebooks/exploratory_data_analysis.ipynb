# Bike Sharing Demand Forecasting - Exploratory Data Analysis (EDA)

## Project Overview

This notebook performs Exploratory Data Analysis (EDA) on the Bike Sharing Dataset.

The goal is to:
- understand rental demand patterns,
- identify key business insights,
- analyze seasonal and weather effects,
- explore relationships between variables,
- and support machine learning forecasting.

Target Variable:

```text
cnt -> Hourly Bike Rental Demand
```


# Import Required Libraries

````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

plt.style.use('ggplot')
sns.set(font_scale=1.0)

print('Libraries imported successfully.')
````

# Define Project Paths

````python
PROJECT_ROOT = Path().resolve().parent

DATA_PATH = (
    PROJECT_ROOT
    / 'data'
    / 'raw'
    / 'hour.csv'
)

GRAPH_PATH = (
    PROJECT_ROOT
    / 'graphs'
)

GRAPH_PATH.mkdir(exist_ok=True)

print(DATA_PATH)
````

# Load Dataset

````python
df = pd.read_csv(DATA_PATH)

print('Dataset Loaded Successfully')

df.head()
````

# Dataset Shape

````python
print('Rows:', df.shape[0])
print('Columns:', df.shape[1])
````

# Dataset Information

````python
df.info()
````

# Statistical Summary

````python
df.describe().T
````

# Check Missing Values

````python
missing_values = df.isnull().sum()

missing_values
````

# Correlation Heatmap

````python
plt.figure(figsize=(14, 10))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(
    correlation_matrix,
    annot=False,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.tight_layout()

plt.savefig(
    GRAPH_PATH / 'correlation_heatmap.png'
)

plt.show()
````

# Hourly Bike Demand Trend

````python
hourly_demand = (
    df.groupby('hr')['cnt']
    .mean()
)

plt.figure(figsize=(12, 6))

hourly_demand.plot(
    marker='o'
)

plt.title('Average Hourly Bike Demand')
plt.xlabel('Hour')
plt.ylabel('Average Demand')

plt.tight_layout()

plt.savefig(
    GRAPH_PATH / 'hourly_demand.png'
)

plt.show()
````

# Seasonal Demand Analysis

````python
seasonal_demand = (
    df.groupby('season')['cnt']
    .mean()
)

plt.figure(figsize=(8, 5))

seasonal_demand.plot(
    kind='bar'
)

plt.title('Seasonal Bike Demand')
plt.xlabel('Season')
plt.ylabel('Average Demand')

plt.tight_layout()

plt.savefig(
    GRAPH_PATH / 'seasonal_trends.png'
)

plt.show()
````

# Weather Impact on Demand

````python
weather_demand = (
    df.groupby('weathersit')['cnt']
    .mean()
)

plt.figure(figsize=(8, 5))

weather_demand.plot(
    kind='bar',
    color='skyblue'
)

plt.title('Weather Impact on Bike Demand')
plt.xlabel('Weather Situation')
plt.ylabel('Average Demand')

plt.tight_layout()

plt.show()
````

# Temperature vs Demand

````python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='temp',
    y='cnt',
    alpha=0.5
)

plt.title('Temperature vs Bike Demand')

plt.tight_layout()

plt.show()
````

# Distribution of Target Variable

````python
plt.figure(figsize=(10, 6))

sns.histplot(
    df['cnt'],
    kde=True,
    bins=40
)

plt.title('Distribution of Bike Rental Demand')

plt.tight_layout()

plt.savefig(
    GRAPH_PATH / 'clv_distribution.png'
)

plt.show()
````

# Working Day vs Holiday Demand

````python
plt.figure(figsize=(10, 5))

sns.boxplot(
    data=df,
    x='workingday',
    y='cnt'
)

plt.title('Working Day vs Bike Demand')

plt.tight_layout()

plt.show()
````

# Peak Hour Analysis

````python
peak_hours = (
    df.groupby('hr')['cnt']
    .mean()
    .sort_values(ascending=False)
)

peak_hours.head(10)
````

# Key Business Insights

## Observations

- Peak bike demand occurs during commuting hours.
- Weather significantly affects bike rentals.
- Summer and Fall have the highest demand.
- Working days show higher commuting usage.
- Temperature positively correlates with bike rentals.
- Demand patterns show strong seasonality.
- Forecasting can significantly improve bicycle logistics.


# Business Recommendations

- Refresh forecasts every 1-3 hours.
- Increase bike availability during commuting hours.
- Monitor weather forecasts for operational planning.
- Retrain models seasonally.
- Use short-term forecasting horizons (24-72 hours).


# Conclusion

The Bike Sharing Dataset demonstrates strong temporal, seasonal, and weather-related demand patterns.

Exploratory Data Analysis shows that:
- hourly usage trends,
- weather conditions,
- seasonality,
- and working day patterns

are highly important for forecasting bike rental demand.

These insights support the development of accurate machine learning forecasting models such as XGBoost for operational bicycle demand prediction.
